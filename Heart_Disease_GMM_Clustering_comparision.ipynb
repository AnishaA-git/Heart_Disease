## **K-Means Clustering Graphs for Heart Disease** 

## Business Logic

With the dataset of Heart Disease, I am using "Age" attribute with other attributes to analyze which age group is more prone towards Heart Disease. So in the dataset we have a total of 14 columns, each column is a feature data which directly helps in predicting the heart disease existence. Considering the "target" column as the prediction result, I have used unsupervised clustering of K-Means to figure out different clusters of 2 different columns. 
Such as:

*   Age : Cholesterol
*   Age : Resting Blood Pressure
*   Age : Max Heart Beats
*   Age : ST Depression

## Objective

To analyze which Age group is more prone for a Heart Disease.

# Cluster Calculation

## Elbow Method for calculating Clusters

```python
def elbowMethod(col1,col2):
  cs = []
  best_cols=[col1.name,col2.name]
  best_vals = df[best_cols].iloc[ :, [0,1]].values
  cs = []
  for i in range(1, 12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    model = kmeans.fit_predict(best_vals)
    #print("model\n", model)
    cs.append(kmeans.inertia_)
  plt.plot(range(1, 12), cs)
  plt.title('The Elbow Method')
  plt.xlabel('Number of clusters')
  plt.ylabel('CS')
  plt.show()
```

## Silhouette Score

```python
def calcSilhouetteScore_GMM(col1,col2):
  best_cols=[col1.name,col2.name]
  best_vals = df[best_cols].iloc[ :, [0,1]].values
  fig, axis = plt.subplots(1,2, figsize = (15,5))
  plt.subplots_adjust(hspace = 0.5,wspace = 0.5)
  cs = []
  for i in range(2, 12):
    gmm_full = GaussianMixture(n_components=i,max_iter = 5, n_init = 5, random_state = 0,covariance_type='full')
    gmm_tied = GaussianMixture(n_components=i,max_iter = 5, n_init = 5, random_state = 0,covariance_type='tied')

    #gmm.fit(df)
    labels_full = gmm_full.fit_predict(best_vals)
    labels_tied = gmm_tied.fit_predict(best_vals)

  #print("model\n", model)
  score_full=metrics.silhouette_score(best_vals, labels_full)
  score_tied=metrics.silhouette_score(best_vals, labels_tied)

    #cs.append(sil)
  scores_full = metrics.silhouette_samples(best_vals, labels_full)
  print("Silhouette Full Score", score_full)
  sns.distplot(scores_full,ax=axis[0]).set(title="Silhouette Score Full")

  scores_tied = metrics.silhouette_samples(best_vals, labels_tied)
  #print("Silhouette Tied Score", scores_tied)
  sns.distplot(scores_tied,ax=axis[1]).set(title="Silhouette Score Tied")
  # plt.plot(range(2, 12), cs)
  # plt.title('The silhouette score')
  # plt.xlabel('Number of clusters')
  # plt.ylabel('score')
  # plt.show()
```

```python
def calcSilhouetteScore_kmeans(col1,col2):
  best_cols=[col1.name,col2.name]
  best_vals = df[best_cols].iloc[ :, [0,1]].values
  cs_kmeans=[]
  for i in range(2, 12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    model = kmeans.fit_predict(best_vals)
    sil=metrics.silhouette_score(best_vals, model, metric='euclidean')
    #print("model\n", model)
    cs_kmeans.append(sil)
  print("The silhouette score",sil)
  plt.plot(range(2, 12), cs_kmeans)
  plt.title('The silhouette score')
  plt.xlabel('Number of clusters')
  plt.ylabel('score')
  plt.show()
```

## Dendogram

```python
def calcDendogram(col1,col2):
  best_cols=[col1.name,col2.name]
  best_vals = df[best_cols].iloc[ :, [0,1]].values
  plt.figure(figsize=(10, 7))
  plt.title("Dendogram")
  dend = shc.dendrogram(shc.linkage(best_vals, method='ward'))
  
```

## **Code Snippet for Scikit Clustering**

```python
def k_means_clustering_scikit(col1,col2,n):
  start=datetime.now()
  print("starttime",start)

  best_cols=[col1.name,col2.name]
  kmeans = KMeans(n_clusters=n, init ="k-means++", n_init=10, max_iter=300)
  best_vals = df[best_cols].iloc[ :, [0,1]].values
  
  print(datetime.now()-start)
  y_pred = kmeans.fit_predict(best_vals)
  print(datetime.now()-start)

  df["K_cluster"] = y_pred
  kmeans.cluster_centers_

  best_cols.append("K_cluster")
  # Number of points in clusters
  print("Number of points in clusters:\n", df['K_cluster'].value_counts())

  # score = metrics.silhouette_score(best_vals, y_pred,metric='euclidean')
  # print("Silhouette score", score)

  # scores = metrics.silhouette_samples(best_vals, y_pred)
  # sns.distplot(scores).set(title="Silhouette Score")

  # score = adjusted_rand_score(best_vals[:,-1], y_pred) 
  # print("adjusted_rand_score", score)

  sns.lmplot(data=df, x=col1.name, y=col2.name, hue='K_cluster', fit_reg=False, legend=True, legend_out=True).set(title="K-Means Cluster Plot")
  plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black',marker='*',label='centroid')
```

## **Code Snippet for GMM**

```python
def GMM_clustering_scikit(col1,col2,n):
  start=datetime.now()

  # fig, axis = plt.subplots(1,2, figsize = (15,5))
  # plt.subplots_adjust(hspace = 0.5,wspace = 0.5)

  best_cols=[col1.name,col2.name]
  gmm_full = GaussianMixture(n_components=n, covariance_type='full',max_iter = 300, n_init = 10)
  gmm_tied = GaussianMixture(n_components=n, covariance_type='tied',max_iter = 300, n_init = 10)

  best_vals = df[best_cols].iloc[ :, [0,1]].values

  print("gmm_full Startime",datetime.now()-start)
  labels_full = gmm_full.fit_predict(best_vals)
  print(datetime.now()-start)

  print("gmm_tied Startime",datetime.now()-start)
  labels_tied = gmm_tied.fit_predict(best_vals)
  print(datetime.now()-start)

  df["GMM_Full_cluster"] = labels_full
  best_cols.append("GMM_Full_cluster")

  df["GMM_Tied_cluster"] = labels_tied
  best_cols.append("GMM_Tied_cluster")


  # Number of points in clusters
  print("Number of points in clusters:\n", df['GMM_Full_cluster'].value_counts())
  gmm_full.means_

  # Number of points in clusters
  print("Number of points in clusters:\n", df['GMM_Tied_cluster'].value_counts())
  gmm_tied.means_

  # score_full = metrics.silhouette_score(best_vals, labels_full)
  # print("Silhouette Full Score", score_full)

  # score_tied = metrics.silhouette_score(best_vals, labels_tied)
  # print("Silhouette Tied Score", score_tied)

  # scores_full = metrics.silhouette_samples(best_vals, labels_full)
  # sns.distplot(scores_full,ax=axis[0]).set(title="Silhouette Full Score")

  # scores_tied = metrics.silhouette_samples(best_vals, labels_tied)
  # sns.distplot(scores_tied,ax=axis[1]).set(title="Silhouette Tied Score")

  # score_full_adj = adjusted_rand_score(best_vals[:,-1], labels_full) 
  # print("adjusted_rand_score Full", score_full_adj) 

  # score_tied_adj = adjusted_rand_score(best_vals[:,-1], labels_tied) 
  # print("adjusted_rand_score Tied", score_tied_adj) 

  sns.lmplot(data=df, x=col1.name, y=col2.name, hue='GMM_Full_cluster', fit_reg=False, legend=True, legend_out=True).set(title="GMM_Full Cluster Plot")
  plt.scatter(gmm_full.means_[:,0],gmm_full.means_[:,1],color='black',marker='*',label='centroid')

  sns.lmplot(data=df, x=col1.name, y=col2.name, hue='GMM_Tied_cluster', fit_reg=False, legend=True, legend_out=True).set(title="GMM_Tied Cluster Plot")
  plt.scatter(gmm_tied.means_[:,0],gmm_tied.means_[:,1],color='black',marker='*',label='centroid')
```

## **Code Snippet for Agglomerative Clustering**

```python
def Agglomerative_clustering_scikit(col1,col2,n):

  start=datetime.now()
  print("starttime",start)

  # fig, axis = plt.subplots(1,3, figsize = (15,5))
  # plt.subplots_adjust(hspace = 0.5,wspace = 0.5)
  best_cols=[col1.name,col2.name]
  agg_ward = AgglomerativeClustering(n_clusters=n, linkage='ward',affinity='euclidean')
  agg_average = AgglomerativeClustering(n_clusters=n, linkage='average',affinity='cityblock')
  agg_complete = AgglomerativeClustering(n_clusters=n, linkage='complete',affinity='cityblock')

  best_vals = df[best_cols].iloc[ :, [0,1]].values

  print(datetime.now()-start)
  y_pred_ward = agg_ward.fit_predict(best_vals)
  y_pred_average = agg_average.fit_predict(best_vals)
  y_pred_complete = agg_complete.fit_predict(best_vals)
  print(datetime.now()-start)

  df["Agglomerative_cluster_ward"] = y_pred_ward
  df["Agglomerative_cluster_average"] = y_pred_average
  df["Agglomerative_cluster_complete"] = y_pred_complete

  best_cols.append("Agglomerative_cluster_ward")
  best_cols.append("Agglomerative_cluster_average")
  best_cols.append("Agglomerative_cluster_complete")

  # Number of points in clusters
  print("Number of points in clusters:\n", df['Agglomerative_cluster_ward'].value_counts())
  # silhouette_scores1 = metrics.silhouette_score(best_vals, y_pred_ward,metric='euclidean')
  # print("Silhouette Score Ward", silhouette_scores1)
  # samples_score1 = metrics.silhouette_samples(best_vals, y_pred_ward)
  # sns.distplot(samples_score1,ax=axis[0]).set(title="Silhouette Score Ward")
  # adjusted_score1 = adjusted_rand_score(best_vals[:,-1], y_pred_ward) 
  # print("adjusted_rand_score", adjusted_score1)

  # Number of points in clusters
  print("Number of points in clusters:\n", df['Agglomerative_cluster_average'].value_counts())
  # silhouette_scores2 = metrics.silhouette_score(best_vals, y_pred_average,metric='euclidean')
  # print("Silhouette Score Average", silhouette_scores2)
  # samples_score2 = metrics.silhouette_samples(best_vals, y_pred_average)
  # sns.distplot(samples_score2,ax=axis[1]).set(title="Silhouette Score Average")
  # adjusted_score2 = adjusted_rand_score(best_vals[:,-1], y_pred_average) 
  # print("adjusted_rand_score", adjusted_score2) 

  # Number of points in clusters
  print("Number of points in clusters:\n", df['Agglomerative_cluster_complete'].value_counts())
  # silhouette_scores3 = metrics.silhouette_score(best_vals, y_pred_complete,metric='euclidean')
  # print("Silhouette Score Complete", silhouette_scores3)
  # samples_score3 = metrics.silhouette_samples(best_vals, y_pred_complete)
  # sns.distplot(samples_score3,ax=axis[2]).set(title="Silhouette Score Complete")
  # adjusted_score3 = adjusted_rand_score(best_vals[:,-1], y_pred_complete) 
  # print("adjusted_rand_score", adjusted_score3) 

  sns.lmplot(data=df, x=col1.name, y=col2.name, hue='Agglomerative_cluster_ward', fit_reg=False, legend=True, legend_out=True).set(title="Agglomerative_cluster_ward")
  sns.lmplot(data=df, x=col1.name, y=col2.name, hue='Agglomerative_cluster_average', fit_reg=False, legend=True, legend_out=True).set(title="Agglomerative_cluster_average")
  sns.lmplot(data=df, x=col1.name, y=col2.name, hue='Agglomerative_cluster_complete', fit_reg=False, legend=True, legend_out=True).set(title="Agglomerative_cluster_complete")
  
```
